### 4. Create a loyalty flag for customers

```python
#Create new column with the maximum order number for users

opm['max_order'] = opm.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Check result

opm.head(10)
```

```python
#Set flags for column

opm.loc[opm['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
opm.loc[(opm['max_order'] <=40) & (opm['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
opm.loc[opm['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
opm['loyalty_flag'].value_counts(dropna = False)
```

### 5. Find any differences in the spending habits of the customer classes

```python
opm.groupby(['loyalty_flag']).agg({'prices':['mean','min','max']})
```

The data would indicate that the newer the customer, the more they spend. This could be due to a number of factors: loyal customers may learn how to navigate deals, or perhaps they order more often and need fewer items per use (leaving high priced items to be purchased more sparingly in each order), or maybe even their familiarity with the platform means there are certain items they do not trust others to purchase for them.

### 6. Create a spending flag for each user based on average price

```python
#Create new column for average price of products per user

opm['average_price'] = opm.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check result

opm.head(15)
```

```python
#Set flags

opm.loc[opm['average_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
opm.loc[opm['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#Check results

opm['spending_flag'].value_counts(dropna = False)
```

```python
opm.head()
```

### 7. Create an order frequency flag

```python
#Create new column for median frequency of use

opm['median_frequency'] = opm.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Check results

opm.head(20)
```

```python
#Create frequency flag column

opm.loc[opm['median_frequency'] > 20, 'user_frequency'] = 'Non-Frequent Customer'
```

```python
opm.loc[(opm['median_frequency'] > 10) & (opm['median_frequency'] <= 20), 'user_frequency'] = 'Regular Customer'
```

```python
opm.loc[opm['median_frequency'] <= 10, 'user_frequency'] = 'Frequent Customer'
```

```python
#Check results

opm['user_frequency'].value_counts(dropna = False)
```

```python
#Create subset to view NaN values for further assessment

opm_nan = opm[opm['user_frequency'].isnull() == True]
opm_nan
```

```python
#View all user 159838's order IDs

opm[opm['user_id'] == 159838]
```

```python
#Add new frequency flag for handling NaN

opm.loc[opm['median_frequency'].isnull() == True, 'user_frequency'] = 'One-Time Customer'
```

```python
#Recheck value counts

opm['user_frequency'].value_counts(dropna = False)
```
